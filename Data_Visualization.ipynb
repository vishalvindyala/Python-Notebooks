```python
# 2. Import data from HW4.csv into data frame, and print the data.
"""
# Age: customer's age in completed years
Education: 1=undergrad; 2=grad; 3=advanced/professional
Experience: years of professional experience
Family: family size
Zip: home address zip code
Income: annual income in $,000
Mortgage: value of house mortgage if any in $,000
Credit: average monthly spending on credit cards in $,000
Loan: Does the customer have a personal loan?
Securities: Does the customer have a securities account?
Deposit: Does the customer have a certificate of deposit account?
Online: Does the customer use internet banking?
Card: Does the customer use a credit card?
"""
import pandas as pd
df = pd.read_csv("HW4.csv")
df
# [5000 rows x 13 columns]
```

```python
# 3. Reset display option to all rows.
# 3.1 Report all unique ages in the ascending order.
# 3.2 Report all descriptive statistics on ages.
pd.set_option('display.max_rows', None)
print(sorted(df["Age"].unique()))
df["Age"].describe()
```

```python
# 4. Reset display option to all available columns.
# 4.1 Report the data, where five columns Loan, Securities, Deposit, Online and Card are all 1.
# [10 rows x 13 columns]
pd.set_option('display.max_columns', None)
df.query("Loan == 1 & Securities == 1 & Deposit == 1 & Online == 1 & Card == 1")
```

```python
# 5. For each year of professional experience, what are the averages per year per column.
# [47 rows x 12 columns]
x = df.groupby(["Experience"],as_index=False).mean()
x
```

```python
# 6. For each combination among five columns Loan, Securities, Deposit, Online and Card, how many records per combination?
# 6.1 After counting by any column, rename that column header to "Count".
# [22 rows x 6 columns]
comb = df.groupby(["Loan","Securities","Deposit","Online","Card"],as_index=False)
comb = comb.agg({"Age":"count"})
comb = comb.rename(columns = {"Age": "Count"})
comb

```

```python
# 7. Report frequency counts on Family sizes.
# 7.1 Hide the grid lines.
# 7.2 Set the family size ticks in the range between 1 and maximum family size to show only integers.
# 7.3 Label the family size axis.
# 7.4 Label the other axis as Number of Customers.
# 7.5 Label the figure title as Customer verssus Family.
import matplotlib.pyplot as plt
family = df["Family"]
n, bins, patches = plt.hist(family)
plt.grid(False)
plt.xticks(range(1,5))
plt.xlabel("Family size")
plt.ylabel("Number of Customers")
plt.title("Customer versues Family")
plt.show()
```

```python
# 8. Report three subplots on one 15"x3" figure as follows:
# 8.1 On the left, report distribution on Income with proper labels and title.
# 8.2 On the center, report distribution on Mortgage with proper labels and title.
# 8.3 On the right, report distribution on Credit with proper labels and title.
plt.figure(figsize=(15,3))
plt.subplot(131)
plt.hist(df["Income"])
plt.xlabel("Income")
plt.ylabel("Number of Custmers")
plt.title("Customer versus Income")
plt.subplot(132)
plt.hist(df["Mortgage"])
plt.xlabel("Mortgage")
plt.ylabel("Number of Customers")
plt.title("Customer versus Mortgage")
plt.subplot(133)
plt.hist(df['Credit'])
plt.xlabel("Credit Sizes")
plt.ylabel("Number of Customer")
plt.title("Customer versus Credit")
plt.show()
```

```python
# 9. Report five pie subplots on one 15"x3" figure as follows:
# 9.1 From left to right, report distribution on Loan, Securities, Deposit, Online and Card, each with proper title.
# 9.2 Each pie subplot should be labeled with 0 and 1 on two colors.
plt.figure(figsize=(15,3))
explode = (0.1,0)
plt.subplot(151)
pie_loan=[df[df['Loan']==1].Loan.count(),df[df['Loan']==0].Loan.count()]
plt.pie(pie_loan,labels=[1,0],explode=explode)
plt.title("Distribution on Loan")

plt.subplot(152)
pie_securities=[df[df['Securities']==1]['Securities'].count(),df[df['Securities']==0]['Securities'].count()]
plt.pie(pie_securities,labels=[1,0],explode=explode)
plt.title("Distribution on Securities")

plt.subplot(153)
pie_deposits=[df[df['Deposit']==1].Deposit.count(),df[df['Deposit']==0].Deposit.count()]
plt.pie(pie_deposits,labels=[1,0],explode=explode)
plt.title("Distribution on Deposit")

plt.subplot(154)
pie_online=[df[df['Online']==1].Online.count(),df[df['Online']==0].Online.count()]
plt.pie(pie_online,labels=[1,0],explode=explode)
plt.title("Distribution on Online")

plt.subplot(155)
pie_card=[df[df['Card']==1].Card.count(),df[df['Card']==0].Card.count()]
plt.pie(pie_card,labels=[1,0],explode=explode)
plt.title("Distribution on Card")
plt.show()
```

```python
# 10. Report numbers of customers for each education level then for each zip code using a treemap.
import plotly.express as px
val = df.groupby(["Education","Zip"],as_index=False)
val = val.agg({"Age":"count"})
val = val.rename(columns = {"Age": "No of Customers"})
fig = px.treemap(val, path=['Education', 'Zip'],values=val["No of Customers"])
fig.show()
```
